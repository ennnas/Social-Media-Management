```python
%matplotlib inline
import matplotlib
import random
import sys
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

single_file_path = "slr12.xls"
multiple_file_path = "mlr02.xls"

def f_costo(X,y,theta):
    size=y.size
    prediction = np.dot(X,theta)
    sqError = (prediction - np.array([y]))**2
    J = ((0.5/size)*sqError.sum())
    return J

def minimize(X,y,theta,size,num_it):
    a=0.00000001 #learning rate
    J_history=[]
    for i in range(num_it):
        J_history.append(f_costo(X,y,theta))
        old_theta=theta
        predictions = np.dot(X,theta)
        for i in range(X.shape[1]):
            grad_x=((predictions - np.array([y])) * X[:, i])
            theta[i]=old_theta[i]-a*(1/size)*grad_x.sum()
    return theta,J_history
        
```
